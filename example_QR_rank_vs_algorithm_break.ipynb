```python
%matplotlib inline
# NB: per eseguire questo notebook come file Python, commentare l'istruzione "%matplotlib inline" qui sopra e scegliere il menÃ¹ "File -> Download as -> Python (.py)"
import numpy as np
import matplotlib.pyplot as plt
import qr 
```

```python
n = 100
[U,_] = np.linalg.qr(np.random.randn(n,n))
[V,_] = np.linalg.qr(np.random.randn(n,n))
S = np.diag(1./np.arange(1,n+1))
A = U @ S @ V.T;
print("cond(A) = ",np.linalg.cond(A))
print("rank(A) = ",np.linalg.matrix_rank(A))
Q, R = qr.MGS(A.copy(),verbose=True)
print("||A - Q R|| = ", np.linalg.norm(A - Q@R,2))
```

```python
k = 10
A[:,[k]] = A[:,0:k]@np.ones((k,1))
print("cond(A) = ",np.linalg.cond(A))
Q, R = qr.MGS(A.copy(),verbose=True)
print("||A - Q R|| = ", np.linalg.norm(A - Q@R,2))
```
